```python
def MSE(pred, true, axis=None):
    return np.mean((pred - true) ** 2, axis=axis)


def MAE(pred, true, axis=None):
    return np.mean(np.abs(pred - true), axis=axis)
```

```python
plt.figure(figsize=(15,3))
mse_err = MSE(pred, true, axis=(1,2))
mae_err = MAE(pred, true, axis=(1,2))
plt.plot(mse_err)
plt.grid(which='both')
plt.ylabel('MSE')

plt.figure(figsize=(15,3))
plt.plot(mae_err)
plt.grid(which='both')
plt.ylabel('MAE')
```
